```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file1 = 'ashrae-energy-prediction/building_metadata.csv'
file2 = 'ashrae-energy-prediction/train.csv'
file3 = 'ashrae-energy-prediction/weather_train.csv'

building_metadata_df = pd.read_csv(file1)
building_train_df = pd.read_csv(file2)
weather_train_df = pd.read_csv(file3)
```

```python
building_train_df['timestamp'] = pd.to_datetime(building_train_df['timestamp'])
building_train_df.head()
```

```python
building_train_df.to_csv('ashrae-energy-prediction/train_clean.csv',index=False)
```

```python
sites = building_metadata_df['site_id'].unique().tolist()

buildings = {}
for site in sites:
    buildings[site] = building_metadata_df[building_metadata_df['site_id']==site]['building_id'].unique().tolist()

temp_df = building_train_df[building_train_df['building_id'] <= max(buildings[0])]
temp_df = temp_df[temp_df['meter_reading'] < max(temp_df['meter_reading'].tolist())]

plt.figure(figsize=(40,20))
sns.lineplot(data=temp_df, x='timestamp', y='meter_reading', hue='building_id')

plt.show();
```
